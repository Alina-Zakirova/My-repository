```python
import pandas as pd 
data = pd.read_csv(r'C:\Users\Admin\Downloads\ratings.csv', sep=",", encoding="utf-8")
movies = pd.read_csv(r"C:\Users\Admin\Downloads\movies.csv", sep=",", encoding="utf-8")

result = pd.merge(data, movies, on="movieId")

def classification(stroka):
    row = stroka['rating']

    if int(row) <= 2:
        return 'Низкий'
    if 2 < int(row) <= 4:
        return "Средний"
    if int(row) >= 4.5:
        return "Высокий"

result["class"] = result.apply(classification, axis=1)
result
```

```python
import pandas as pd 
data = pd.read_csv(r'C:\Users\Admin\Downloads\keywords.csv', sep=",", encoding="utf-8")

geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

def classify_region(query):
    query_lower = query.lower() 
    for region, cities in geo_data.items():
        for city in cities:
            if city in query_lower:
                return region 
    return 'undefined'  

data['region'] = data['keyword'].apply(classify_region)

region_counts = data['region'].value_counts()
print(region_counts)

data.tail()
```
